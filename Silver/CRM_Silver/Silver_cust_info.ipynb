## Cleaning steps

1. Remove duplicates
2. Remove all row nulls
3. Fill the nulls/''/' '  
4. Trim spaces
5. Standardize case
6. Cast data types
7. Filter invalid rows
7. Rename columns
9. Add metadata columns if applicable
10. Validate schema

```python
import pyspark.sql.functions as F
from pyspark.sql.types import StringType
```

## Load table

```python
df = spark.table('bikes_catalog.bronze.bronze_cust_info')
df.show()
```

## Cleaning

```python
df = df.dropDuplicates()
```

```python
df = df.dropna(how='all')
```

```python
for col in df.columns:
    print(f'Nulls in {col}: {df.filter(F.col(col).isNull()).count()}')
```

```python
df.filter(df.cst_id.isNull()).show()
```

```python
df = df.na.drop(thresh = 2)
```

```python
df.filter(df.cst_firstname.isNull()).show()
```

```python
for col in df.columns[2:6]:
    df = df.na.fill({f'{col}': 'Unknown'})
```

```python
for col in df.columns:
    print(f'Nulls in {col}: {df.filter(F.col(col).isNull()).count()}')
```

```python
for field in df.schema.fields:
    if isinstance(field.dataType, StringType):
        df = df.withColumn(field.name,F.trim(F.col(field.name)))

df.show()
```

```python
for col in df.columns[4:6]:
    print(f'Distinct values in {col}: {df.select(F.col(col)).distinct().count()}')
    print(df.select(F.col(col)).distinct().show())
```

```python
for old_names, new_names in zip(df.columns, ['C_key','CID','Firstname','Lastname','Marital_status','Gender','Cst_creation_date']):
    df = df.withColumnRenamed(old_names, new_names)
df.show()
```

```python
df.printSchema()
```

```python
df.write.mode('overwrite').saveAsTable('bikes_catalog.silver.silver_cust_info')
```
