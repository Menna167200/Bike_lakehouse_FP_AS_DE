```python
df = spark.sql('''
        SELECT 
        ROW_NUMBER() OVER(ORDER BY p.Product_key) As Product_sk,
        p.Product_key,
        p.Product_name,
        c.Category_ID,
        c.Category,
        c.Sub_category,
        p.Product_line, 
        p.Product_cost,
        c.Maintainance,
        p.Product_start_date,
        p.Product_end_date

        FROM bikes_catalog.silver.silver_prd_info AS p
        LEFT JOIN bikes_catalog.silver.silver_px_cat_g1v2 AS c
        ON p.Category_ID = c.Category_ID
          ''')

df.show()
```

```python
df.write.mode('overwrite').saveAsTable('bikes_catalog.gold.gold_product')
```
