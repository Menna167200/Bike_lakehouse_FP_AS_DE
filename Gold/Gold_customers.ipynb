```python
df = spark.sql('''
        SELECT 
        c.C_key AS Customer_key, --key
        ca.CID AS Customer_ID,
        c.Firstname,
        c.Lastname,
        c.Gender,
        c.Marital_status,
        cl.Country,
        ca.BDATE AS Birthdate,
        c.Cst_creation_date

        FROM bikes_catalog.silver.silver_cust_info AS c
        LEFT JOIN bikes_catalog.silver.silver_cust_az12 AS ca ON c.CID = ca.CID
        LEFT JOIN bikes_catalog.silver.silver_loc_a101 AS cl ON c.CID = cl.CID
          ''')
df.show()
```

```python
df.write.mode('overwrite').saveAsTable('bikes_catalog.gold.gold_customer')
```
