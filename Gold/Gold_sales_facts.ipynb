```python
df = spark.sql('''
        SELECT 
        s.Order_number,
        p.Product_key,
        c.Customer_key,
        s.Order_date,
        s.Ship_date,
        s.Due_date,
        s.Sales,
        s.Quantity,
        s.Price
        
        FROM bikes_catalog.silver.silver_sales_details AS s
        LEFT JOIN bikes_catalog.gold.gold_product AS p ON s.Product_key = p.Product_key
        LEFT JOIN bikes_catalog.gold.gold_customer AS c ON s.C_key = c.Customer_key
          ''')

df.show()
```

```python
df.write.mode('overwrite').saveAsTable('bikes_catalog.gold.gold_sales')
```
