## Set the env

```python
%sql
USE CATALOG bikes_catalog;
USE SCHEMA bronze;
```

```python
files = {'source_crm' : ['cust_info','prd_info','sales_details'],
         'source_erp': ['CUST_AZ12','LOC_A101','PX_CAT_G1V2']}

path = '/Volumes/bikes_catalog/bronze/source_data/' 
```

## Read/Write

```python
for folder,files in files.items():
    for file in files:
        df = spark.read.csv(f'{path}{folder}/{file}.csv', header=True, inferSchema=True)
        df.write.mode('overwrite').saveAsTable(f'bronze_{file}')
```
